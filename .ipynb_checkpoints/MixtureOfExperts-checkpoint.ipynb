```python
# TensorFlow and tf.keras
import tensorflow as tf
from tensorflow import keras

# Helper libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


print(tf.__version__)
```

```python
df = pd.read_csv('full_results.csv', index_col='dataid')
df_features = df[['ARIMA', 'Holt-Winters', 'TESLA', 'LSTM-20-7']]
df_labels = df[['Actual']]

df_features['ARIMA'] = df_features['ARIMA'].apply(eval)
df_features['Holt-Winters'] = df_features['Holt-Winters'].apply(eval)
df_features['TESLA'] = df_features['TESLA'].apply(eval)
df_features['LSTM-20-7'] = df_features['LSTM-20-7'].apply(eval)

df_labels['Actual'] = df_labels['Actual'].apply(eval)

total_features = df_features.values.tolist()
total_labels = df_labels.values.tolist()

for x in range(len(total_labels)) :
    while len(total_labels[x][0]) < 7027 :
        total_labels[x][0].append(0)
        
for x in range(len(total_features)) :
    for y in range(len(total_features[x])) :
        while len(total_features[x][y]) < 7027 :
            total_features[x][y].append(0)

new_total_features = np.array(total_features, order=2)
new_total_labels = np.array(total_labels, order=3)


train_houses = new_total_features[:18]
test_houses = new_total_features[18:]
train_labels = new_total_labels[:18]
test_labels = new_total_labels[18:]


df_training = pd.read_csv('trainingMeanData.csv', index_col='dataid')

training_list = df_training.values
training_list.shape

gate_training = training_list[:18]
gate_testing = training_list[18:]
```

```python
model = keras.Sequential([
    keras.layers.Dense(17, input_shape=(17,), activation='softmax'),
#     keras.layers.Dense(20, activation='softmax'),
#     keras.layers.Dense(20, activation='softmax'),
    keras.layers.Dense(100, activation='softmax'),
    keras.layers.Dense(100, activation='softmax'),
    keras.layers.Dense(200, activation='softmax'),
    keras.layers.Dense(100, activation='softmax'),
    keras.layers.Dense(100, activation='softmax'),
    keras.layers.Dense(4, activation='softmax')
])
```

```python
from sklearn.metrics import mean_squared_error
import keras.backend as K

def test_loss(features, train) :
    
    def custom_loss(y_actual, y_predicted) :
        total = 0
        tf.print(y_predicted[0])
        for x in range(features.shape[0]) :    
            
            new_arima = curr_total_features[x][0] * y_predicted[0][0]
            new_holt = curr_total_features[x][1] * y_predicted[0][1]
            new_tesla = curr_total_features[x][2] * y_predicted[0][2]
            new_lstm = curr_total_features[x][3] * y_predicted[0][3]
    
            new = tf.math.add(new_arima, new_holt)
            new = tf.math.add(new, new_tesla)
            new = tf.math.add(new, new_lstm)
            
            total = total + K.mean(K.square(new - train[x]))


        return total / features.shape[0]
    
    return custom_loss

curr_total_features = tf.Variable(train_houses, dtype=tf.float32)
labels = train_labels
model.compile(optimizer='adam',
              loss=test_loss(curr_total_features, labels))
```

```python
model.fit(gate_training, train_labels, epochs=5000)
```

```python
curr_total_features = tf.Variable(test_houses, dtype=tf.float32)
labels = test_labels
model.compile(optimizer='adam',
              loss=test_loss(curr_total_features, labels))
test_loss = model.evaluate(gate_testing,  test_labels, verbose=2)

print('\nTest loss:', test_loss)
```
